# **Concat (pd.concat()) :-**

```python
import pandas as pd
import numpy as np

jan = pd.DataFrame({
    "Product":["Laptop","Mouse"],
    "Sales":[50000,500]
})

feb = pd.DataFrame({
    "Product":["Keyboard","Monitor"],
    "Sales":[1500,12000]
})

#1. Row-wise Concat (Most Common)
print("Row-wise Concat:\n",pd.concat([jan,feb]))

#2. Column-wise Concat
print("\nColumn-wise Concat:\n",pd.concat([jan,feb],axis=1))

#3. Concat with Ignore Index
print("\nConcat with Ignore Index:\n",pd.concat(
    [jan, feb],
    ignore_index=True
))

#4. Concat with Keys
print("\nConcat with Keys:\n",pd.concat(
    [jan, feb],
    keys=["January","February"]
))
```

# **Merge(pd.merge()) :-**

```python
import pandas as pd
import numpy as np

employees = pd.DataFrame({
    "Emp_ID":[1,2,3,4],
    "Name":["Rahul","Priya","Amit","Neha"],
    "Dept_ID":[101,102,105,103]
})

departments = pd.DataFrame({
    "Dept_ID":[101,102,104],
    "Department":["IT","HR","Finance"]
})


#Beginner Tasks

#Q1 : Perform Inner Join.
inner_join=pd.merge(employees,departments,on="Dept_ID",how="inner")
print("Perform Inner Join :\n",inner_join)
print("--------------------------------------------------")

#Q2 : Perform Left Join.
left_join=pd.merge(employees,departments,on="Dept_ID",how="left")
print("\nPerform Left Join :\n",left_join)
print("--------------------------------------------------")


#Q3 : Perform Right Join.
right_join=pd.merge(employees,departments,on="Dept_ID",how="right")
print("\nPerform Right Join :\n",right_join)
print("--------------------------------------------------")


#Q4 : Perform Outer Join.
outer_join=pd.merge(employees,departments,on="Dept_ID",how="outer")
print("\nPerform Outer Join :\n",outer_join)
print("--------------------------------------------------")


#Q5 : Which employees don't have a matching department?
employees_without_department = pd.merge(
    employees,
    departments,
    on="Dept_ID",
    how="outer",
    indicator=True
)
print("Which employees don't have a matching department? :\n",employees_without_department[employees_without_department["_merge"] == "left_only"][["Emp_ID", "Name"]])
print("--------------------------------------------------")
#[employees_without_department["_merge"] == "left_only"] means : Give me only those rows that exist in the employees table but not in the departments table.


#Intermediate Tasks

#Q6 : Use indicator=True.
indicator_join=pd.merge(employees,departments,on="Dept_ID",how="outer",indicator=True)
print("\nUse indicator=True :\n",indicator_join)
print("--------------------------------------------------")


#Q7 : Count rows in each merge type.
merge_counts = pd.merge(
    employees,
    departments,
    on="Dept_ID",
    how="outer",
    indicator=True
)
merge_type_counts = merge_counts["_merge"].value_counts()
print("\nCount rows in each merge type :\n",merge_type_counts)
print("--------------------------------------------------")


#Q8 : Find departments without employees.
df=pd.merge(employees,departments,on="Dept_ID",how="outer",indicator=True)
departments_without_employees=df[df["_merge"] =="right_only"][["Department","Dept_ID"]]
print("\nFind departments without employees :\n",departments_without_employees)
print("--------------------------------------------------")

#Q9 : Find employees whose department is missing.
df=pd.merge(employees,departments,on="Dept_ID",how="outer",indicator=True)
departments_missing_employees=df[df["_merge"] =="left_only"]["Name"]
print("\nFind employees whose department is missing :\n",departments_missing_employees)
print("--------------------------------------------------")

#Q10 : Find matching records only.
print("\nFind matching records only :\n",pd.merge(employees,departments,on="Dept_ID",how="inner"))
print("--------------------------------------------------")
```

# **Join(.join()) :-**

```python
#Beginner
#Q1 : Create two DataFrames with the same index and join them.
import pandas as pd

employees = pd.DataFrame({
    "Name": ["Rahul", "Priya", "Amit"]
}, index=[101, 102, 103])

print("Employees DataSet :\n",employees)

salary = pd.DataFrame({
    "Salary": [50000, 45000]
}, index=[101, 102])

print("\nSalary DataSet :\n",salary)

#Q2 : Perform a left join.
left_join=employees.join(salary,how="left")
print("\nPerform a left join :\n",left_join)

#Q3 : Perform an inner join.
inner_join=employees.join(salary,how="inner")
print("\nPerform an inner join :\n",inner_join)

#Q4 : Perform an outer join.
outer_join=employees.join(salary,how="outer")
print("\nPerform an outer join :\n",outer_join)
```
