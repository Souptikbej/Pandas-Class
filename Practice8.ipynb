# **Concat (pd.concat()) :-**

```python
import pandas as pd
import numpy as np

jan = pd.DataFrame({
    "Product":["Laptop","Mouse"],
    "Sales":[50000,500]
})

feb = pd.DataFrame({
    "Product":["Keyboard","Monitor"],
    "Sales":[1500,12000]
})

#1. Row-wise Concat (Most Common)
print("Row-wise Concat:\n",pd.concat([jan,feb]))

#2. Column-wise Concat
print("\nColumn-wise Concat:\n",pd.concat([jan,feb],axis=1))

#3. Concat with Ignore Index
print("\nConcat with Ignore Index:\n",pd.concat(
    [jan, feb],
    ignore_index=True
))

#4. Concat with Keys
print("\nConcat with Keys:\n",pd.concat(
    [jan, feb],
    keys=["January","February"]
))
```
