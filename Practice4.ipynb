# Project 1

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "Name":["Rahul","Priya","Rahul"," Amit ","Neha"],
    "Age":[25,np.nan,25,28,np.nan],
    "Salary":[50000,45000,50000,np.nan,55000]
})

print(df.info())
print("____________________________________")
print("Raw Data :\n",df)
print("____________________________________")

# Find missing values.
print("Missing Value :\n",df.isnull())
print("____________________________________")
# Count missing values.
print("Count Missing Value :\n",df.isnull().sum())
print("____________________________________")
# Fill Age with mean.
df["Age"].fillna(df["Age"].mean(),inplace=True)
print("Fill Age with mean :\n",df)
print("____________________________________")
# Fill Salary with median.
df['Salary'].fillna(df['Salary'].median(),inplace=True)
print("Fill Salary with median :\n",df)
print("____________________________________")
# Remove duplicates.
df.drop_duplicates(inplace=True)
print("Remove duplicates :\n",df)
print("____________________________________")
# Remove spaces from Name.
df['Name']=df['Name'].str.strip()
print("Remove spaces from Name :\n",df)
print("____________________________________")
# Convert Name to uppercase.
df['Name']=df['Name'].str.upper()
print("Convert Name to uppercase :\n",df)
print("____________________________________")
# Check data types.
print("Data types :\n",df.dtypes)
```

## **Level 1: Basic Cleaning**

**Missing Values**

1. Find missing values in the dataset.
1. Count missing values column-wise.
1. Find total missing values in the full dataset.
1. Find rows where Name is missing.
1. Fill missing Name with "Unknown".
1. Fill missing Age with the median age.
1. Fill missing Salary with the median salary.
1. Fill missing Gender with "Unknown".
1. Fill missing Join_Date with "Not Provided".
1. Replace empty string in Department with "Unknown".

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "Customer_ID": [101, 102, 103, 104, 105, 105, 106, 107],
    "Name": [" Rahul ", "PRIYA", "amit", "Neha ", "rohan", "rohan", None, "Anjali"],
    "Age": [25, np.nan, 28, 27, 30, 30, -5, 26],
    "Gender": ["Male", "Female", "male", "FEMALE", "M", "M", "Female", None],
    "City": ["Kolkata", "kolkata ", "Delhi", "DELHI", "Mumbai", "Mumbai", "Pune", "pune"],
    "Salary": [50000, 45000, np.nan, 55000, 70000, 70000, 40000, "not available"],
    "Join_Date": ["2024-01-10", "10/02/2024", "2024-03-15", None, "2024-05-20", "2024-05-20", "wrong date", "2024-07-01"],
    "Department": ["IT", "HR", "it", "Finance", "IT Dept", "IT Dept", "HR", ""]
})

print(df.dtypes)

#Level 1: Basic Cleaning

#Missing Values

#Find missing values in the dataset.
print("Missing Values :\n",df.isna())
print("-------------------------------------")
#Count missing values column-wise.
print("Count missing values column-wise :\n",df.isna().sum())
print("-------------------------------------")
#Find total missing values in the full dataset.
print("Find total missing values in the full dataset : ",df.isna().sum().sum())
print("-------------------------------------")
#Find rows where Name is missing.
print("Find rows where Name is missing :\n",df[df['Name'].isna()])
print("-------------------------------------")
#Fill missing Name with "Unknown".
df['Name']=df['Name'].fillna("Unknown")
print("Fill missing Name with 'Unknown' :\n",df)
print("-------------------------------------")
#Fill missing Age with the median age.
df['Age']=df['Age'].fillna(df['Age'].median())
print("Fill missing Age with the median age :\n",df)
print("-------------------------------------")
#Fill missing Salary with the median salary.
df['Salary']=df['Salary'].replace("not available",np.nan)
df['Salary']=df['Salary'].fillna(df['Salary'].median())
print("Fill missing Salary with the median salary :\n",df)
print("-------------------------------------")
#Fill missing Gender with "Unknown".
df['Gender']=df['Gender'].fillna("Unknown")
print("Fill missing Gender with 'Unknown' :\n",df)
print("-------------------------------------")
#Fill missing Join_Date with "Not Provided".
df['Join_Date']=df['Join_Date'].fillna("Not Provided")
print("Fill missing Join_Date with 'Not Provided' :\n",df)
print("-------------------------------------")
#Replace empty string in Department with "Unknown".
df['Department']=df['Department'].replace("","Unknown")
print("Replace empty string in Department with 'Unknown' :\n",df)
print("-------------------------------------")
```

## **Level 2: Duplicate Cleaning**
1. Check duplicate rows.
1. Count duplicate rows.
1. Remove duplicate rows.
1. Check duplicates based only on Customer_ID.
1. Remove duplicate records based on Customer_ID.

```python
#Level 2: Duplicate Cleaning
#Check duplicate rows.
print("Check Duplicate Rows :\n",df.duplicated())
print("-------------------------------------")
#Count duplicate rows.
print("Count of Duplicate Rows is : ",df.duplicated().sum())
print("-------------------------------------")
#Remove duplicate rows.
df.drop_duplicates(inplace=True)
print("Remove Duplicate Rows :\n",df)
print("-------------------------------------")
#Check duplicates based only on Customer_ID.
print("Check duplicates based only on Customer_ID. :\n",df.duplicated(subset="Customer_ID"))
print("-------------------------------------")
#Remove duplicate records based on Customer_ID.
df.drop_duplicates(subset="Customer_ID",inplace=True)
print("Remove duplicate records based on Customer_ID :\n",df)
print("-------------------------------------")
```

## **Level 3: Text Cleaning**
1. Remove extra spaces from Name.
1. Convert Name to title case. Example:
`"rahul" → "Rahul"`
1. Convert City to title case.
1. Remove extra spaces from City.
1. Standardize Gender. Expected:
`Male, Female, Unknown`
1. Convert M to Male.
1. Convert male to Male.
1. Convert FEMALE to Female.
1. Standardize Department. Expected:
`IT, HR, Finance, Unknown`
1. Convert IT Dept and it to IT.

```python
#Remove extra spaces from Name.
df['Name']=df['Name'].str.strip()
#Convert Name to title case. Example: "rahul" → "Rahul"
df['Name']=df['Name'].str.title()
#Convert City to title case.
df['City']=df['City'].str.title()
#Remove extra spaces from City.
df['City']=df['City'].str.strip()
#Standardize Gender. Expected: Male, Female, Unknown
df['Gender']=df['Gender'].str.title()
df['Gender']=df['Gender'].replace(["Male","Female","Unknown"],["Male","Female","Unknown"])
#Convert M to Male.
df['Gender']=df['Gender'].replace("M","Male")
#Convert male to Male.
df['Gender']=df['Gender'].replace("male","Male")
#Convert FEMALE to Female.
df['Gender']=df['Gender'].replace("FEMALE","Female")
#Standardize Department. Expected: IT, HR, Finance, Unknown
df['Department']=df['Department'].str.title()
df['Department']=df['Department'].replace(["It Dept","Finance","Hr"],["IT","Finance","HR"])
#Convert IT Dept and it to IT.
df['Department']=df['Department'].replace(["IT Dept","It"],"IT")
print(df)
#
```

## **Level 4: Data Type Cleaning**
1. Check data types.
1. Convert Salary to numeric.
1. Convert invalid salary values to NaN.
1. Fill invalid salary values with median salary.
1. Convert Age to integer.
1. Convert Join_Date to datetime.
1. Convert invalid dates to NaT.
1. Extract Join_Year.
1. Extract Join_Month.
1. Extract Join_Day.

```python
#Check data types.
print(df.dtypes)
print("-------------------------------------")
#Convert Salary to numeric.
df['Salary']=df['Salary'].astype(float)
print("After Convert Salary to numeric :\n",df.dtypes)
print("-------------------------------------")
#Convert invalid salary values to NaN.
#Fill invalid salary values with median salary.
#Convert Age to integer.
df['Age']=df['Age'].astype(int)
print("Convert Age to integer :\n",df.dtypes)
print("-------------------------------------")
#Convert Join_Date to datetime.
df['Join_Date']=pd.to_datetime(df['Join_Date'],errors='coerce')
print("Convert Join_Date to datetime :\n",df.dtypes)
print("-------------------------------------")
#Convert invalid dates to NaT.
df['Join_Date']=df['Join_Date'].replace("wrong date",pd.NaT)
print("Convert invalid dates to NaT :\n",df)
print("-------------------------------------")
#Extract Join_Year.
#Extract Join_Month.
#Extract Join_Day.
df['Join_Year']=df['Join_Date'].dt.year
df['Join_Month']=df['Join_Date'].dt.month
df['Join_Day']=df['Join_Date'].dt.day
print("Extract Join_Year ,Join_Month , Join_Day :\n",df)
```
