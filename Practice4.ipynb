```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "Name":["Rahul","Priya","Rahul"," Amit ","Neha"],
    "Age":[25,np.nan,25,28,np.nan],
    "Salary":[50000,45000,50000,np.nan,55000]
})

print(df.info())
print("____________________________________")
print("Raw Data :\n",df)
print("____________________________________")

# Find missing values.
print("Missing Value :\n",df.isnull())
print("____________________________________")
# Count missing values.
print("Count Missing Value :\n",df.isnull().sum())
print("____________________________________")
# Fill Age with mean.
df["Age"].fillna(df["Age"].mean(),inplace=True)
print("Fill Age with mean :\n",df)
print("____________________________________")
# Fill Salary with median.
df['Salary'].fillna(df['Salary'].median(),inplace=True)
print("Fill Salary with median :\n",df)
print("____________________________________")
# Remove duplicates.
# Remove spaces from Name.
# Convert Name to uppercase.
# Check data types.
```
