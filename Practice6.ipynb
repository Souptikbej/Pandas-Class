````python
#Sorting
import pandas as pd
import numpy as np
df = pd.DataFrame({
    "Name":["Rahul","Priya","Amit","Neha","Rohan"],
    "Age":[25,24,28,27,30],
    "Salary":[50000,45000,60000,55000,70000]
})

print("----------------------------")
print("Original DataFrame:")
print(df)
print("----------------------------")

#Sort Salary ascending.
df.sort_values(by="Salary",ascending=True,inplace=True)
print("Sorted by Salary ascending:")
print(df)
print("----------------------------")

#Sort Salary descending.
df.sort_values(by="Salary",ascending=False,inplace=True)
print("Sorted by Salary descending:")
print(df)
print("----------------------------")
#Sort Age descending.
df.sort_values(by="Age",ascending=False,inplace=True)
print("Sorted by Age descending:")
print(df)
print("----------------------------")
#Find top 3 highest salaries.
print("Top 3 highest salaries:")
print(df.nlargest(3,"Salary"))
print("----------------------------")
````

````python
print("Original DataFrame:")
print(df)
print("----------------------------")
#Average salary.
print("Average salary:")
print(df["Salary"].mean())
print("----------------------------")
#Median salary.
print("Median salary:")
print(df["Salary"].median())
print("----------------------------")
#Maximum salary.
print("Maximum salary:")
print(df["Salary"].max())
print("----------------------------")
#Minimum salary.
print("Minimum salary:")
print(df["Salary"].min())
print("----------------------------")
#Total salary.
print("Total salary:")
print(df["Salary"].sum())
print("----------------------------")
#Standard deviation of salary.
print("Standard deviation of salary:")
print(df["Salary"].std())
print("----------------------------")
````

````python
#Value Counts
#Count unique ages.
print("Count unique ages:")
print(df["Age"].nunique())
print("----------------------------")
#Count frequency of ages.
print("Count frequency of ages:")
print(df["Age"].value_counts())
print("----------------------------")
#Find most common age.
print("Most common age:")
print(df["Age"].mode())#Can't find most common age because all values are come one time
print("----------------------------")
````

````python
#Quantiles
#Calculate Q1.
print("Q1:")
print(df["Salary"].quantile(0.25))
print("----------------------------")
#Calculate Median.
print("Median:")
print(df["Salary"].median())
print("----------------------------")
#Calculate Q3.
print("Q3:")
print(df["Salary"].quantile(0.75))
print("----------------------------")
````
