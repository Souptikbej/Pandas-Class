# **Advanced DataTime**

````python
import pandas as pd

df = pd.DataFrame({
    "Order_ID":[101,102,103,104,105],
    "Customer":["Rahul","Priya","Amit","Neha","Rohan"],
    "Order_Date":[
        "2024-01-15",
        "2024-02-20",
        "2024-03-12",
        "2024-03-18",
        "2024-04-25"
    ],
    "Delivery_Date":[
        "2024-01-18",
        "2024-02-24",
        "2024-03-16",
        "2024-03-22",
        "2024-04-30"
    ],
    "Sales":[5000,7000,4500,9000,8000]
})

df["Order_Date"] = pd.to_datetime(df["Order_Date"])
df["Delivery_Date"] = pd.to_datetime(df["Delivery_Date"])


##1. .dt.year
#How many orders were placed in each year?
print("How many orders were placed in each year? :\n",df["Order_Date"].dt.year.value_counts())
print("---------------------------------------------------")

#What is the total sales for each year?
print("What is the total sales for each year? :\n",df.groupby(df["Order_Date"].dt.year)["Sales"].sum())
print("---------------------------------------------------")

#Show only orders placed in 2024.
print("Show only orders placed in 2024. :\n",df[df["Order_Date"].dt.year == 2024])
print("---------------------------------------------------")


##2. .dt.month
#Which month generated the highest sales?
print("Which month generated the highest sales? :\n",df["Order_Date"].dt.month.value_counts())
print("---------------------------------------------------")

#What is the total sales for each month?
print("What is the total sales for each month? :\n",df.groupby(df["Order_Date"].dt.month)["Sales"].sum())
print("---------------------------------------------------")

#How many orders came in March?
print("How many orders came in March? :\n",df[df["Order_Date"].dt.month == 3])
print("---------------------------------------------------")


#3. .dt.day
#On which day of the month do customers place the most orders?
print("On which day of the month do customers place the most orders? :\n",df["Order_Date"].dt.day.value_counts())
print("---------------------------------------------------")

#Show all orders placed on the 15th.
print("Show all orders placed on the 15th. :\n",df[df["Order_Date"].dt.day == 15])
print("---------------------------------------------------")


#4. .dt.day_name()
#Which weekday receives the highest number of orders?
print("Which weekday receives the highest number of orders? :\n",df["Order_Date"].dt.day_name().value_counts())
print("---------------------------------------------------")

#How many orders were placed on Mondays?
print("How many orders were placed on Mondays? :\n",df[df["Order_Date"].dt.day_name() == "Monday"])
print("---------------------------------------------------")

#What is the total sales on Mondays?
print("What is the total sales on Mondays? :\n",df[df["Order_Date"].dt.day_name() == "Monday"]["Sales"].sum())
print("---------------------------------------------------")
````

## **5. .dt.month_name()**

````python
#Which month has the highest revenue?

df.groupby(
    df["Order_Date"].dt.month_name()
)["Sales"].sum()
````

## **6. .dt.weekday**


```
Monday    = 0

```



````python
#How many weekday vs weekend orders do we receive?

print(df["Order_Date"].dt.weekday.value_counts())
````

## **7. .dt.quarter**

````python
#Which quarter generated the highest sales?

print("Which quarter generated the highest sales?\n",df.groupby(df["Order_Date"].dt.quarter)["Sales"].sum())
print("---------------------------------------------------")

#How many orders were placed in Q1?

print("How many orders were placed in Q1?\n",df[df["Order_Date"].dt.quarter == 1])
print("---------------------------------------------------")
````

## **8. .dt.isocalendar().week**

````python
#Which week had the highest number of orders?

print("Which week had the highest number of orders?\n",df.groupby(df["Order_Date"].dt.isocalendar().week).size())
print("---------------------------------------------------")


#Weekly sales trend.
print("Weekly sales trend\n",df.groupby(df["Order_Date"].dt.isocalendar().week)["Sales"].sum())
````

## **9. Date Filtering**

````python
#Show all March orders.
print("Show all March orders.\n",df[df["Order_Date"].dt.month == 3])
print("---------------------------------------------------")

#Show orders after March.
print("Show orders after March.\n",df[df["Order_Date"] > "2024-03-01"])
print("---------------------------------------------------")


#Orders between January and March.
print("Orders between January and March.\n",df[
    (df["Order_Date"] >= "2024-01-01") &
    (df["Order_Date"] <= "2024-03-31")
])
````

## **10. Date Difference**

````python
#How many days did delivery take?
df["Delivery_Time"] = (df["Delivery_Date"] -df["Order_Date"])
print("How many days did delivery take?\n",df["Delivery_Time"])
print("---------------------------------------------------")


#Which order had the longest delivery time?
print("Which order had the longest delivery time?\n",df.sort_values("Delivery_Time",ascending=False))
print("---------------------------------------------------")
````

## **12. pd.Timestamp.now()**

````python
#How many days ago was the order placed?

df["Days_Since_Order"] = ( pd.Timestamp.now() - df["Order_Date"]).dt.days
print("How many days ago was the order placed?\n",df["Days_Since_Order"])
print("---------------------------------------------------")

#Customers inactive for more than 180 days.
print("Customers inactive for more than 180 days.\n",df[df["Days_Since_Order"] > 180])
print("---------------------------------------------------")
````

## **13. pd.date_range()**

````python
#Create a calendar for all dates in January.
print("Create a calendar for all dates in January.\n",pd.date_range("2024-01-01","2024-01-31"))
print("---------------------------------------------------")

#Generate monthly report dates.

print("Generate monthly report dates.\n",pd.date_range(start="2024-01-01",end="2024-12-31",freq="ME"))
````

## **14. Resampling (Very Important)**

**Imagine sales every day.**

````python
#Monthly sales report.
print("Monthly sales report.\n",df.resample("ME")["Sales"].sum())
print("---------------------------------------------------")

#Quarterly sales.
print("Quarterly sales.\n",df.resample("QE")["Sales"].sum())
print("---------------------------------------------------")


#Yearly sales.
print("Yearly sales.\n",df.resample("YE")["Sales"].sum())
print("---------------------------------------------------")
````

## **15. Rolling Window**

````python
#What is the 7-day moving average of sales?
print("What is the 7-day moving average of sales?\n",df["Sales"].rolling(7).mean())
print("---------------------------------------------------")
````

## **16. Expanding Window**

````python
#What is the cumulative average sales over time?
print("What is the cumulative average sales over time?\n",df["Sales"].expanding().mean())
print("---------------------------------------------------")
````
