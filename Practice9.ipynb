# **Advanced DataTime**

```python
import pandas as pd

df = pd.DataFrame({
    "Order_ID":[101,102,103,104,105],
    "Customer":["Rahul","Priya","Amit","Neha","Rohan"],
    "Order_Date":[
        "2024-01-15",
        "2024-02-20",
        "2024-03-12",
        "2024-03-18",
        "2024-04-25"
    ],
    "Delivery_Date":[
        "2024-01-18",
        "2024-02-24",
        "2024-03-16",
        "2024-03-22",
        "2024-04-30"
    ],
    "Sales":[5000,7000,4500,9000,8000]
})

df["Order_Date"] = pd.to_datetime(df["Order_Date"])
df["Delivery_Date"] = pd.to_datetime(df["Delivery_Date"])


##1. .dt.year
#How many orders were placed in each year?
print("How many orders were placed in each year? :\n",df["Order_Date"].dt.year.value_counts())
print("---------------------------------------------------")

#What is the total sales for each year?
print("What is the total sales for each year? :\n",df.groupby(df["Order_Date"].dt.year)["Sales"].sum())
print("---------------------------------------------------")

#Show only orders placed in 2024.
print("Show only orders placed in 2024. :\n",df[df["Order_Date"].dt.year == 2024])
print("---------------------------------------------------")


##2. .dt.month
#Which month generated the highest sales?
print("Which month generated the highest sales? :\n",df["Order_Date"].dt.month.value_counts())
print("---------------------------------------------------")

#What is the total sales for each month?
print("What is the total sales for each month? :\n",df.groupby(df["Order_Date"].dt.month)["Sales"].sum())
print("---------------------------------------------------")

#How many orders came in March?
print("How many orders came in March? :\n",df[df["Order_Date"].dt.month == 3])
print("---------------------------------------------------")


#3. .dt.day
#On which day of the month do customers place the most orders?
print("On which day of the month do customers place the most orders? :\n",df["Order_Date"].dt.day.value_counts())
print("---------------------------------------------------")

#Show all orders placed on the 15th.
print("Show all orders placed on the 15th. :\n",df[df["Order_Date"].dt.day == 15])
print("---------------------------------------------------")


#4. .dt.day_name()
#Which weekday receives the highest number of orders?
print("Which weekday receives the highest number of orders? :\n",df["Order_Date"].dt.day_name().value_counts())
print("---------------------------------------------------")

#How many orders were placed on Mondays?
print("How many orders were placed on Mondays? :\n",df[df["Order_Date"].dt.day_name() == "Monday"])
print("---------------------------------------------------")

#What is the total sales on Mondays?
print("What is the total sales on Mondays? :\n",df[df["Order_Date"].dt.day_name() == "Monday"]["Sales"].sum())
print("---------------------------------------------------")
```
