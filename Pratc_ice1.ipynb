```python
import pandas as pd

students = {
    "Name": ["Rahul", "Priya", "Amit", "Neha"],
    "Age": [20, 21, 22, 20],
    "Marks": [85, 92, 78, 88]
}

df = pd.DataFrame(students)

print(df)


#Show only Name column.
print(df["Name"])


#Show Name and Marks.
print(df[["Name","Marks"]])


#Add Grade column.
df["Grade"] = ["B","A","C","B"]
print(df)


#Check shape.
df.shape

#Use head()
df.head()

#Use describe().
df.describe()

```
