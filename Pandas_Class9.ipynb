## 1. Applying Custom Functions with `apply()` / `map()`
- `apply()` lets you apply a custom function across rows or columns.
- `map()` is for applying functions element-wise to a Series.
- Both are handy for transformations, calculations, and cleaning data.

```python
import pandas as pd

df = pd.DataFrame({
    "Employee":["Rahul","Priya","Amit","Neha"],
    "Salary":[30000,60000,90000,120000],
    "Age":[18,24,35,45]
})
print("Data Set :\n",df)
print("------------------------------------")

#Manager asks: Categorize salaries (Below 50k → Low) , (50k–80k → Medium) , (Above 80k → High)
def cate(n):
  if n<50000:
    return "Low"
  elif n>=50000 and n<=80000:
    return "Medium"
  else:
    return "High"
df["Salaries_Categorize"]=df['Salary'].apply(cate)
print("Categorized Data Set :\n",df)
print("------------------------------------")


#Manager asks: Give every employee a 10% bonus.
df["Bonus"]=df['Salary'].apply(lambda x: x*0.1)
print("Giving 10% bonus for every employee :\n",df)
print("------------------------------------")

#Classify customers (Teen,Adult,Senior)
def age_s(n):
  if n<18:
    return "Teen"
  elif n>=18 and n<=40:
    return "Adult"
  else:
    return "Senior"
df["Age_Categorize"]=df['Age'].apply(age_s)
print("Categorized Age Data Set :\n",df)
print("-------------------------------------------------")

#Convert every employee name to uppercase.
df['Employee']=df['Employee'].apply(lambda x: x.upper())
print("Uppercase Employee Names :\n",df)
print("-------------------------------------------------")

#Manager asks: Generate company email IDs.
df['Email']=df['Employee'].apply(lambda s: s.lower()+"@company.com")
print("Company Email IDs :\n",df)
print("--------------------------------------------------")

#Tax Calculation : Salary Rules (Under 50k → 5%), (50k–100k → 10%), (Above 100k → 20%)
df['Tax']=df['Salary'].apply(lambda x: x*0.05 if x<50000 else(x*0.1 if x>=50000 and x<=100000 else x*0.2))
print("Tax Calculation :\n",df)
print("--------------------------------------------------")
```
