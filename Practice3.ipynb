# **Data Filtering**

# **Tasks**

1. Show only Name column.
1. Show Name and Salary.
1. Show first 3 rows using iloc.
1. Show rows 1–4 using loc.
1. Find salary > 50,000.
1. Find Department = IT.
1. Find IT employees with salary > 55,000.
1. Find HR or Finance employees.
1. Count employees in each department.
1. Find unique departments.

```python
import pandas as pd
import numpy as np

employees = {
    "Name":["Rahul","Priya","Amit","Neha","Rohan","Anjali"],
    "Department":["IT","HR","IT","Finance","IT","HR"],
    "Salary":[50000,45000,60000,55000,70000,48000],
    "Age":[25,24,28,27,30,26]
}

df = pd.DataFrame(employees)
print(df)
print("___________________________")

#Show only Name column.
print(df["Name"])
print("___________________________")
#Show Name and Salary.
print(df[["Name","Salary"]])
print("___________________________")
#Show first 3 rows using iloc.
print(df.iloc[0:4])
print("___________________________")
#Show rows 1–4 using loc.
print(df.iloc[1:4])
print("___________________________")
#Find salary > 50,000.
print(df[df['Salary']>50000])
print("___________________________")
#Find Department = IT.
print(df[df['Department']=='IT'])
print("___________________________")
#Find IT employees with salary > 55,000.
print(df[(df['Department']=='IT') & (df['Salary']>55000)])
print("___________________________")
#Find HR or Finance employees.
print(df[(df['Department']=='HR') | (df['Department']=='Finance')])
print("___________________________")
#Count employees in each department.
print(df['Department'].value_counts())
print("___________________________")
#Find unique departments.
print(df['Department'].unique())
print("___________________________")
```
