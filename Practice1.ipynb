````python
import pandas as pd

students = {
    "Name": ["Rahul", "Priya", "Amit", "Neha"],
    "Age": [20, 21, 22, 20],
    "Marks": [85, 92, 78, 88]
}

df = pd.DataFrame(students)
print("Raw Data :")
print(df)
print("___________________________")


#Show only Name column.
print("\n1. Show only Name column : ")
print(df["Name"])
print("___________________________")

#Show Name and Marks.
print("\n2. Show Name and Marks : ")
print(df[["Name","Marks"]])
print("___________________________")

#Add Grade column.
df["Grade"] = ["B","A","C","B"]
print("\n3. Add Grade column :")
print(df)
print("___________________________")

#Check shape.
print("\n4. Shape is : \n",df.shape)
print("___________________________")

#Check data types.
print("\n4. Data types are : \n",df.dtypes)
print("___________________________")

#Use head()
print("\n5. Use head() : \n",df.head())
print("___________________________")
#Use describe().
print("\n6. Use describe() : \n",df.describe())
````
