```python
import pandas as pd

df = pd.DataFrame({
    "Name":["Rahul","Priya","Amit","Neha","Rohan","Anjali"],
    "Department":["IT","HR","IT","Finance","IT","HR"],
    "City":["Kolkata","Delhi","Kolkata","Mumbai","Delhi","Delhi"],
    "Salary":[50000,45000,60000,55000,70000,48000],
    "Age":[25,24,28,27,30,26]
})

print("Original Data Set -\n",df)
print("------------------------------------")

#1. Basic GroupBy
#Count Employees by Department
print("Count Employees by Department")
print(df.groupby("Department")["Name"].count()) #1 way
print(df.groupby("Department").size())  #2 way
print("------------------------------------")

#2. Mean
#Average Salary by Department
print("2.Average Salary by Department\n",df.groupby("Department")["Salary"].mean())
print("------------------------------------")

#3. Sum
#Total Salary by Department
print("3.Total Salary by Department\n",df.groupby("Department")["Salary"].sum())
print("------------------------------------")

#4. Min and Max
#Highest Salary by Department
print("4.1. Highest Salary by Department\n",df.groupby("Department")["Salary"].max())
print("------------------------------------")

#Lowest Salary by Department
print("4.2. Lowest Salary by Department\n",df.groupby("Department")["Salary"].min())
print("------------------------------------")

#5. Multiple Aggregations
print("5. Multiple Aggregations\n",df.groupby("Department")["Salary"].agg(["count","mean","min","max","sum"]))
print("------------------------------------")

#6. Group By Multiple Columns Department + City
print("6. Group By Multiple Columns Department + City\n",df.groupby(["Department","City"])["Salary"].mean())
print("------------------------------------")
#7. as_index=False
print("7.1. Default--\n",df.groupby("Department",as_index=True)["Salary"].mean())
print("7.2. as_index=False--\n",df.groupby("Department",as_index=False)["Salary"].mean())
print("------------------------------------")

#9. Named Aggregation
print("9. Named Aggregation")
print(df.groupby("Department").agg(Average_Salary=("Salary","mean"),Total_Salary=("Salary","sum")))
print("------------------------------------")


#10. Sort GroupBy Results
print("10. Sort GroupBy Results")
print(df.groupby("Department")["Salary"].mean().sort_values(ascending=False))
print("------------------------------------")



#############Real Business Questions########################
print("Real Business Questions")
print("------------------------------------")
#Which department pays the highest average salary?
print("Which department pays the highest average salary?")
print(df.groupby("Department")["Salary"].mean())
print("------------------------------------")

#Which city has the highest total salary?
print("Which city has the highest total salary?")
print(df.groupby("City")["Salary"].sum())
print("------------------------------------")

#How many employees are in each department?
print("How many employees are in each department?")
print(df.groupby("Department").size())
print("------------------------------------")

#Highest-paid employee per department?
print("Highest-paid employee per department?")
print(df.groupby("Department")["Salary"].max())
print("------------------------------------")
```

```python
import pandas as pd

df = pd.DataFrame({
    "Name":["Rahul","Priya","Amit","Neha","Rohan","Anjali"],
    "Department":["IT","HR","IT","Finance","IT","HR"],
    "City":["Kolkata","Delhi","Kolkata","Mumbai","Delhi","Delhi"],
    "Salary":[50000,45000,60000,55000,70000,48000],
    "Age":[25,24,28,27,30,26]
})
print("Original Data Set -\n",df)
print("------------------------------------")
#Count employees per department.
print("Count employees per department.\n",df.groupby("Department").size())
print("------------------------------------")
#Count employees per city.
print("Count employees per city.\n",df.groupby("City").size())
print("------------------------------------")
#Average salary by department.
print("Average salary by department.\n",df.groupby("Department")["Salary"].mean())
print("------------------------------------")
#Average age by department.
print("Average age by department.\n",df.groupby("Department")["Age"].mean())
print("------------------------------------")
#Total salary by department.
print("Total salary by department.\n",df.groupby("Department")["Salary"].sum())
print("------------------------------------")
```

```python
#Highest salary by department.
print("Highest salary by department.\n",df.groupby("Department")["Salary"].max())
print("------------------------------------")
#Lowest salary by city.
print("Lowest salary by city.\n",df.groupby("City")["Salary"].min())
print("------------------------------------")
#Salary statistics by department.
print("Salary statistics by department.\n",df.groupby("Department")["Salary"].describe())
print("------------------------------------")
#Average salary by city.
print("Average salary by city.\n",df.groupby("City")["Salary"].mean())
print("------------------------------------")
#Department with highest average salary.
print("Department with highest average salary.\n",df.groupby("Department")["Salary"].mean().idxmax())#The idxmax() method in pandas returns the index label of the first occurrence of the maximum value along a specified axis.
print("------------------------------------")
```
