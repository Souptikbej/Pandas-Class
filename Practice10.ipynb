# **Apply() Practice**

```python
import pandas as pd

df = pd.DataFrame({
    "Employee_ID": [101, 102, 103, 104, 105, 106, 107, 108],
    "Name": [
        "Rahul Sharma",
        "Priya Das",
        "Amit Kumar",
        "Neha Singh",
        "Rohan Roy",
        "Anjali Sen",
        "Sourav Ghosh",
        "Pooja Verma"
    ],
    "Age": [17, 22, 29, 35, 42, 58, 15, 67],
    "Salary": [30000, 55000, 75000, 90000, 120000, 45000, 25000, 150000],
    "Math": [78, 92, 55, 68, 88, 45, 81, 95],
    "Science": [82, 90, 60, 72, 91, 50, 76, 98],
    "English": [75, 89, 58, 70, 85, 48, 79, 96],
    "Sales": [120000, 180000, 95000, 150000, 250000, 110000, 70000, 300000],
    "Cost": [80000, 120000, 70000, 90000, 180000, 75000, 50000, 220000],
    "Product_Price": [500, 1200, 800, 2500, 4500, 700, 300, 6000],
    "Total_Purchase": [8000, 25000, 15000, 50000, 120000, 18000, 5000, 200000]
})

print(df)
```

```python
#Create a Salary Band (Low, Medium, High).
df["Salary_Band"]=df['Salary'].apply(lambda x: "Low" if x<50000 else ("Medium" if x>=50000 and x<=80000 else "High"))
print(df)
```

```python
#Calculate a 15% bonus for every employee.
df["Bouns"]=df['Salary'].apply(lambda x: x*0.15)
print(df)
```

```python
#Generate company email addresses from employee names.
df['Email']=df['Name'].apply(lambda x: x.strip().replace(" ",".").lower() +"@company.com")
print(df)
```

```python
#Categorize ages into Child, Adult, and Senior.
df['Age_Group']=df['Age'].apply(lambda x: "Child" if x<18 else ("Adult" if x>=18 and x<=60 else "Senior"))
print(df)
```

```python
#Calculate Total Marks from three subject columns.
df['Total_Marks']=df[['Math','Science','English']].apply(lambda x: x.sum(),axis=1)
print(df)
```

```python
#Create a Profit column (Sales - Cost).
df['Profit']=df[['Sales','Cost']].apply(lambda x: x['Sales']-x['Cost'],axis=1)
print(df)
```

```python
#Create a Pass/Fail column based on marks.
df['Result']=df['Total_Marks'].apply(lambda x: "Pass" if x>=180 else "Fail")
print(df)
```

```python
#Generate a unique employee code like EMP101, EMP102, etc.
df['Employee_Code']=df['Employee_ID'].apply(lambda x: "EMP"+str(x))
print(df)
```

```python
#Apply a discount based on product price.
df['Discount']=df['Product_Price'].apply(lambda x: x*0.05 if x<1000 else(x*0.1 if x>=1000 and x<=3000 else x*1.5))
print(df)
```

```python
#Assign customer loyalty levels based on total purchases.
df['Loyalty_Level']=df['Total_Purchase'].apply(lambda x: "Bronze" if x<10000 else ("Silver" if x>=10000 and x<=50000 else "Gold"))
print(df)
```
