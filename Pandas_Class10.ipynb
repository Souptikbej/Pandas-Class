## **What is Feature Engineering?**
Feature engineering means creating new, meaningful features (columns) from existing data to improve analysis or model performance.


````python
import pandas as pd

df = pd.DataFrame({
    "Employee_ID": [101,102,103,104,105],
    "Name": ["Rahul Sharma","Priya Sen","Amit Ghosh","Neha Dhuly","Rohan Dhara"],
    "Age": [24,30,45,27,38],
    "Salary": [40000,75000,120000,55000,90000],
    "Sales": [120000,180000,250000,150000,210000],
    "Cost": [80000,120000,180000,100000,140000],
    "Experience": [2,5,15,4,10],
    "Join_Date": pd.to_datetime([
        "2022-01-10",
        "2020-05-15",
        "2015-08-20",
        "2021-03-01",
        "2018-11-12"
    ])
})
df
````

## **Feature 1: Profit**
**Business Question**
> What profit did each employee generate?





````python
df['Profit']=df['Sales']-df['Cost']
df
````

## **Feature 2: Profit Margin %**
**Business Question**

>Which employee has the highest profit margin?

````python
df['Profit_margin']=(df['Profit']/df['Sales'])*100
df
````

## **Feature 3: Experience Level**
**Business Question**

>Categorize employees by experience.


```
0–2 years      → Beginner
3–7 years      → Intermediate
8+ years       → Expert
```



````python
def exp(n):
  if n<=2:
    return 'Beginner'
  elif n<=7 and n>=3:
    return 'Intermediate'
  else:
    return 'Expert'
df['Experience_Level']=df['Experience'].apply(exp)
df
````

## **Feature 4: Salary Band**
**Business Question**

>Group employees into salary bands.
```
< 50000      → Low
50000–90000  → Medium
> 90000      → High
```

````python
def band(n):
  if n<50000:
    return 'Low'
  elif n<=90000 and n>=50000:
    return 'Medium'
  else:
    return 'High'
df['Salary_Band']=df['Salary'].apply(band)
df
````

## **Feature 5: Employee Tenure**
**Business Question**

>How many days has each employee worked?

````python
df['Tenure_Days'] = (pd.Timestamp.today() - df['Join_Date']).dt.days
df
````

## **Feature 6: Employee Tenure (Years)**
**Business Question**

>How many years has each employee worked?

````python
df['Tenure_Years'] = (df['Tenure_Days']/365).round(1)
df
````

## **Feature 7: Revenue per Year of Experience**
**Business Question**

>Which employee generates the most sales relative to experience?

````python
df["Sales_Per_Year"] = (df["Sales"] / df["Experience"]).round()
df["Sales_Per_Year"] = df["Sales_Per_Year"].astype(int)
df
````

## **Feature 8: High Performer**
**Business Question**

>Mark employees as High Performer if profit exceeds ₹60,000.

````python
df["High_Performer"] = df["Profit"].apply(lambda x: "Yes" if x > 60000 else "No")
df
````

## **Feature 9: Initials**
**Business Question**

>Generate employee initials.
```
Rahul Sharma
↓
RS
```

````python
df["Initials"] = df["Name"].apply(lambda x: "".join(word[0] for word in x.split()))
df
````

## **Feature 10: Employee Email**
**Business Question**

>Create company email IDs.
```
Rahul
↓
rahul@company.com

```

````python
df['Email']=df['Name'].apply(lambda x: x.lower().replace(" ","")+"@company.com")
df
````
