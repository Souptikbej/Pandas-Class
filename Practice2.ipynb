*Create a CSV file:*



```
Name,Department,Salary
Rahul,IT,50000

Neha,Finance,55000
```

**Task 1**

Load file
`pd.read_csv()`

**Task 2**

`head(3)`

**Task 3**

Save as:
`employee_backup.csv`


````python
import pandas as pd
from google.colab import drive

# Mount Google Drive
drive.mount('/content/drive')

# Read CSV
df = pd.read_csv('/content/drive/MyDrive/employees.csv')
print(df)
print("_________________________________")
print(df.info())
print("_________________________________")
#All columns name
print(df.columns)
print("_________________________________")
#Find the total number of rows and columns
print(df.shape)
print("_________________________________")
# Preview data
print(df.head(3))
print("_________________________________")
#Display the last 2 rows
print(df.tail(2))
print("_________________________________")
# Save backup file
df.to_csv('/content/drive/MyDrive/employee_backup.csv', index=False)

print("Backup saved successfully!")
````
